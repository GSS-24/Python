# Guía sobre print() en Python

## 1. Ejercicios básicos.

### Mostrar texto.

Debes colocar el texto entre comillas dentro de paréntesis.

```python
print("Hola, mundo")
```

```python
print("Bienvenido a Python")
```

```python
print("¡Esto es muy fácil!")
```

### Mostrar números.

Los número se colocaán sin comillas.

```python
print(42)
```

```python
print(3.14159)
```

Así se muestran el resultado de operaciones.

```python
print(10 +5)
```

```python
print(100 / 4)
```

### Colocar línea vacía.

Para colocar un espacio solo se debe colocar un print() vacío para imprimir una línea en blanco.

```python
print("Primer línea")
print()
print("Segunda línea")
```

## 2. Comillas simples y dobles

### Ambas tienen la misma función.

Las comillas simples o dobles se usan para colocar texto.

```python
print("Hola")
```

```python
print('Hola')
```

### Cuando el texto tiene comillas.

Si tu texto incluye comillas debes usar el otro tipo de para encerrarlo.

```python
print("Él dijo 'hola' y se fue")
```

```python
print('Ella respondió "adiós"')
```

## 3. Imprimir varias cosas a la vez.

### Separar datos con comas.

Si deseas imprimir varios elementos, debes separarlos con comas para que se impriman con un espacio entre cada uno.

```python
print("Hola", "mundo")
```

```python
print("El resultado es:", 42)
```

```python
print("Suma:", 10, "+", 5, "=", 15)
```

### Utilizar función "sep".

La función "sep" por defecto es un espacio, pero se puede sustituir por otros elementos.

##### Separador por defecto (espacio).

```python
print("A", "B", "C")
```

##### Separador personalizado (guión).

```python
print("A", "B", "C", sep="-")
```

##### Sin separador (Sin espacios).

```python
print("A", "B", "C", sep="")
```

##### Separador con diagonal (fechas).

```python
print(2026, 10, 6, sep="/")
```

##### Separador con dos puntos (horas).

```python
print(14, 30, 00, sep=":")
```

## 4. Controlar los saltos de línea.

### Función normal.

Cada print() termina con un salto de línea automático.

```python
print("Línea 1")
print("Línea 2")
print("Línea 3")
```

### Utilizar la función "end".

La función "end" por defecto es un salto de línea, pero se puede sustituir por otros elementos.

##### Mantener todo en la misma línea.

```python
print("Hola", end=" ")
print("mundo")
```

##### Colocar puntos suspensivos.

```python
print("Cargando", end="...")
print("Listo")
```

##### Sin espacios.

```python
print("A", end="")
print("B", end="")
print("C")
```

## 5. Caracteres especiales.

### Secuencias de escape.

Existen caracteres que se imprimen en secuencia que empiezan con "\"

##### \n = Nueve línea

```python
print("Primera línea\nSegunda línea")
```

##### \t = Tabulación

```python
print("Nombre:\tJuan")
print("Edad:\t25")
```

##### \\ = Barra invertida

```python
print("Carpeta: C:\\Usuarios\\Juan")
```

##### \\" Comilla doble

```python
print("Dijo \"hola\"")
```

##### \\' Comilla Simple

```python
print('It\'s ok')
```

## 6. Texto en varias líneas.

### Comillas triples.

Para textos largos que ocupa varias lineas se utilizan tres comillas. (no importa si son dobles """Texto""" o simples '''Texto'''

```python
print("""
Bienvenido al sistema.
Por favor, siga las instrucciones.
Gracias por su visita.
""")
```

### Crear dibujoa con texto.

LAs comillas triples son perfectas para el arte ASCII o menús

```python
print("""
╔═══════════════════════╗
║    MI PROGRAMA        ║
║                       ║
║  1. Opción uno        ║
║  2. Opción dos        ║
║  3. Salir             ║
╚═══════════════════════╝
""")
```
