# Guía sobre print() en Python

## 1. Ejercicios básicos.

### Mostrar texto.

Debes colocar el texto entre comillas dentro de paréntesis.

```python
print("Hola, mundo")
```

```python
print("Bienvenido a Python")
```

```python
print("¡Esto es muy fácil!")
```

### Mostrar números.

Los número se colocaán sin comillas.

```python
print(42)
```

```python
print(3.14159)
```

Así se muestran el resultado de operaciones.

```python
print(10 +5)
```

```python
print(100 / 4)
```

### Colocar línea vacía.

Para colocar un espacio solo se debe colocar un print() vacío para imprimir una línea en blanco.

print("Primer línea")
print()
print("Segunda línea")

## 2. Comillas simples y dobles

### Ambas tienen la misma función.

Las comillas simples o dobles se usan para colocar texto.

```python
print("Hola")
```

```python
print('Hola')
```

### Cuando el texto tiene comillas.

Si tu texto incluye comillas debes usar el otro tipo de para encerrarlo.

```python
print("Él dijo 'hola' y se fue")
```

```python
print('Ella respondió "adiós"')
```

## 3. Imprimir varias cosas a la vez.

### Separar datos con comas.

Si deseas imprimir varios elementos, debes separarlos con comas para que se impriman con un espacio entre cada uno.

```python
print("Hola", "mundo")
```

```python
print("El resultado es:", 42)
```

```python
print("Suma:", 10, "+", 5, "=", 15)
```

### Utilizar función "sep"

La función "sep" por defecto es un espacio, pero se puede sustituir por otros elementos.
